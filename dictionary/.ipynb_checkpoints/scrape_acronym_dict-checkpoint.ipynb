```python
#!/usr/bin/env python

# importing the requests library 
import requests 
import re
import pandas as pd
```

```python
# api-endpoint 
URL = "https://www.noslang.com/dictionary/1/"
  
# sending get request and saving the response as response object 
r = requests.get(url = URL) 
```

```python
html = r.text
```

```python
translations = re.search(r'<abbr(...)+["In case you didn\'t catch it"](...)+>', html)
```

```python
translations.group(0).split('<abbr class="dictonary-abbr" ')
```

```python
acronyms = re.search(r'<dt>(...)+</dt>', html)
```

```python
acronyms.group(0).split('<dt>')
```

# construct dictionary

```python
acronym_dict = dict()
df = pd.DataFrame(columns = ['slang', 'translation', 'raw'])
for appendix in '1abcdefghijklmnopqrstuvwxyz':
    URL = "https://www.noslang.com/dictionary/" + appendix + "/"
    r = requests.get(url = URL)
    html = r.text
    trans_text = re.search(r'<abbr(...)+>', html)
    trans_list = trans_text.group(0).split('<abbr class="dictonary-abbr" ')
    slang_text = re.search(r'<dt>(...)+</dt>', html)
    slang_list = slang_text.group(0).split('<dt>')
    for i in range(len(trans_list)):
        trans_raw = trans_list[i] if i < len(trans_list) else "null"
        slang_raw = slang_list[i] if i < len(slang_list) else "null"
        if trans_raw.find('title') != -1 and slang_raw != '':
            # print(text)
            slang = slang_raw[: slang_raw.find(' :</dt>')] if slang_raw != 'null' else 'null'
            trans = trans_raw[7: trans_raw.find('"><span class')] if trans_raw != 'null' else 'null'
            new_row = pd.DataFrame([[slang, trans, trans_list[i]]], columns=df.columns)
            df = pd.concat([df, new_row], ignore_index=True)
```
